```python
import folium
import geopandas as gpd
# import matplotlib.pyplot as plt

# Carregar os dados geográficos dos municípios de São Paulo
Esp = gpd.read_file(r'C:\Projetos\GEO\fontes\SP_Municipios_2024.zip')

# Filtrar os dados para os municípios de São Paulo e Diadema
sp = Esp[Esp['NM_MUN'] == 'São Paulo']
diadema = Esp[Esp['NM_MUN'] == 'Diadema']
```

```python
# Os geodataframes 'sp' e 'diadema' agora contêm as geometrias dos municípios de São Paulo e Diadema, respectivamente e podem ser usados para plotar no mapa já que possuem a coluna 'geometry' com as geometrias dos municípios na mesma projeção.

# 1. Initialize the map (adjust location/zoom as needed)
# Using a sample location if data not available
m = folium.Map(location=[-23.8475, -46.6361], zoom_start=11, tiles='CartoDB positron')

# 2. Plota primeira coluna de geometria
# Set 'geometry_col1' as active geometry and add to map
sp.set_geometry(sp['geometry'])
# Use the 'explore' method to add to the map.
# Assign a name for the LayerControl
sp.explore(m=m, color='blue', name='São Paulo', marker_kwds={'radius': 5})

# 3. Plota segunda coluna de geometria
# Set 'geometry_col2' as active geometry and add to the SAME map object
diadema.set_geometry(diadema['geometry'])
diadema.explore(m=m, color='red', marker_kwds={'radius': 5}, name='Diadema')

# 4. Adiciona o Controle de Camadas para selecionar aquelas interessantes
# Note: When using gdf.explore(), it automatically adds a LayerControl if you set 'name'
# However, you can explicitly add one if more control is needed or for other folium objects.
folium.LayerControl().add_to(m)


# 5. Exibe o mapa
m
```
